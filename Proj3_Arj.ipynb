###### Importing Libraries

```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

###### Importing Datasets

```python
df_ind_occ  = pd.read_csv(r'https://data.london.gov.uk/download/jobs-by-occupation-and-highest-qualification/b9a48a98-9f3f-4722-8cbc-6be7cc33ffbd/jobs_by_%20industry_and_occupation.csv')
df_occ_qual  = pd.read_csv(r'https://data.london.gov.uk/download/jobs-by-occupation-and-highest-qualification/9f15d50c-f63f-4eb8-a39b-5ba3a5ff5b0d/jobs_by_%20occupation_and_highest_qualification.csv')
```

###### Displaying the head of industry_occupation dataset

```python
df_ind_occ.head()
```

###### Note: str.strip() is used to remove all the leading and trailing spaces from a string.

```python
df_ind_occ.columns = df_ind_occ.columns.str.strip()
```

###### Displaying the head of occupation_qualification dataset

```python
df_occ_qual.head()
```

###### Note: str.strip() is used to remove all the leading and trailing spaces from a string.

```python
df_occ_qual.columns = df_occ_qual.columns.str.strip()
```

###### Merging both the datasets based on "Occupation" column by using "Inner Join"

Reference: https://towardsdatascience.com/why-and-how-to-use-merge-with-pandas-in-python-548600f7e738

```python
df = pd.merge(df_ind_occ,df_occ_qual, on="occupation",how="inner")
df.tail()
```

###### Checking for any existing null values in the dataframe after merging both the datasets.

```python
df.isnull().sum()
```

## Data Analysis

## Hypothesis - 1

### For each highest_degree_qualification occupation "X" has the highest number of jobs and the same occupation has the highest number of jobs in "X" industry.

```python
grouped = df.groupby('highest_educational_qualification')
for key,grp in grouped:
  r_df = grp.loc[grp['occupation'] != 'Total in all occupations'] 
  plt.figure()
  fig = r_df['number_of_jobs_y'].plot()
  fig.set_title("Max No of jobs for Qualification : "+key)
  fig.set_xlabel('No of Jobs')
  max_df = r_df.loc[r_df['number_of_jobs_y'] == r_df['number_of_jobs_y'].max()]
  print("Qualification:"+ key + " | "+ "Occupation with Max jobs: "+str(max_df['occupation'].unique()[0]))
  max_occupation = max_df['occupation'].unique()[0]
  new_df = df_ind_occ[df_ind_occ['occupation'] == max_occupation]
  new_df['number_of_jobs'] = new_df[new_df['number_of_jobs']!=".."]['number_of_jobs'].astype(int) 
  plt.figure()
  fig = new_df['number_of_jobs'].plot()
  fig.set_title("Max No of jobs for Occupation : "+max_occupation)
  fig.set_xlabel('No of Jobs')

  ind_max_df = new_df.loc[new_df['number_of_jobs'] == new_df['number_of_jobs'].max()]
  print("Occupation : "+ max_occupation + " | "+ "Industry with Max jobs: "+str(ind_max_df['industry'].unique()[0]))
  print("****************************************************************************************")
plt.show()
```

## Hypothesis - 2

### For each highest_degree_qualification occupation "X" has the least number of jobs and the same occupation has the least number of jobs in "X" industry.

```python
# Minimum jobs
grouped = df.groupby('highest_educational_qualification')
for key,grp in grouped:
  r_df = grp.loc[grp['occupation'] != 'Total in all occupations'] 
  plt.figure()
  fig = r_df['number_of_jobs_y'].plot()
  fig.set_title("MiN No of jobs for Qualification : "+key)
  fig.set_xlabel('No of Jobs')
  min_df = r_df.loc[r_df['number_of_jobs_y'] == r_df['number_of_jobs_y']. min()]
  print("Qualification:"+ key + " | "+ "Occupation with Min jobs: "+str(min_df['occupation'].unique()[0]))
  min_occupation = min_df['occupation'].unique()[0]
  new_df = df_ind_occ[df_ind_occ['occupation'] == min_occupation]
  new_df['number_of_jobs'] = new_df[new_df['number_of_jobs']!=".."]['number_of_jobs'].astype(int) 
  plt.figure()
  fig = new_df['number_of_jobs'].plot()
  fig.set_title("min No of jobs for Occupation : "+min_occupation)
  fig.set_xlabel('No of Jobs')

  ind_min_df = new_df.loc[new_df['number_of_jobs'] == new_df['number_of_jobs'].min()]
  print("Occupation : "+ min_occupation + " | "+ "Industry with min jobs: "+str(ind_min_df['industry'].unique()[0]))
  print("****************************************************************************************")
  break
plt.show()
```

### "X" year has the highest number of jobs per industry per highest_educational_qualification.

## Hypothesis - 4

### X industry has the highest number of jobs with highest educational qualification as higher degree

```python
df.head()
```

```python
grouped  = df.groupby('industry')

for key,group in grouped:
  r_df = group[group['highest_educational_qualification'] == 'Higher degree']
  max_jobs = r_df.loc[r_df['number_of_jobs_x'] ==r_df['number_of_jobs_x'].max()]
  print("Industry: "+max_jobs['industry'].unique()[0]  + " Number Of Jobs :" + str(max_jobs['number_of_jobs_x'].unique()[0]))
  max_jobs['number_of_jobs_x'] = max_jobs['number_of_jobs_x'].astype(int)
  plt.figure()
  fig = max_jobs['number_of_jobs_x'].plot.pie()
  fig.set_title("Max No of jobs for Industry with qualification as higher degree : "+key)
  fig.set_xlabel('No of Jobs')
  print("************************************************************************************************************")
plt.show()
```

## Hypothesis - 5

### For "X" Educational qualification, Maximum number of Jobs per Industry

```python
df.head()
```

```python
grouped = df.groupby('highest_educational_qualification')
for key,group in grouped:
  plt.figure()
  for ind_key,ind_group in group.groupby('industry'):
    max_jobs = ind_group[ind_group['number_of_jobs_x'] == ind_group['number_of_jobs_x'].max()]
    plt.figure()
    max_jobs['number_of_jobs_x'] = max_jobs['number_of_jobs_x'].astype(int)
    fig = max_jobs['number_of_jobs_x'].plot.pie()
    fig.set_title("Max No of jobs for Industry : "+ind_key)
    fig.set_xlabel('No of Jobs')
    print("*******************************************")
    max_count = max_jobs['number_of_jobs_x'].unique()[0]
    print("Industry: "+ind_key + " : "+"Max jobs : "+str(max_count))
    break
plt.show()
```
